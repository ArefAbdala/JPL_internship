# 2. Installing Jupyter

For those not familiar with Jupyter Notebooks, it is a software that is able to combine code and text for easier documentation of data. The author recommends the installation of the Anaconda package. This package already includes the jupyter notebook software and a plethora of python libraries.





Keep in mind that when accessing Jupyter, there are two options: Jupyter Notebook and Jupyterlab. While both are able to be installed and can be run in the browser, JupyterLab has a wider range of capabilities when compared to it's counterpart. To access JupeterLab, open the Anaconda prompt 

Install JupyterLab with pip:

pip install jupyterlab

Once installed, launch JupyterLab with:

jupyter-lab


# 3. netrc file

Depending on what data you want to access and where, the netrc file will look different. This file should be in the home directory and containyour login information for Earthdata. More inofrmation on how to create this file, can be found here: https://nasa-openscapes.github.io/2021-Cloud-Hackathon/tutorials/04_NASA_Earthdata_Authentication.html#authentication-via-netrc-file



The file should be in the following format:

machine urs.earthdata.nasa.gov <br>
login "username" <br>
password "password"

**notice** the netrc file's format may change when trying to access cloud data
