# MUR- OPeNDAP Access Tutorial

#### Where to find it?
This tutorial's github code can be found here: https://github.com/podaac/tutorials/blob/master/notebooks/opendap/MUR-OPeNDAP.ipynb

## Notebook Objectives:
Search the common metadata repository (CMR) for the MUR dataset<br>
Obtain OPeNDAP links from CMR search<br>
Download data from OPeNDAP links and open via xarray to visualize data


#### Preface
Using this tutorial will require packages listed in this section. The user will need to create an Earth Data login and a corresponding netrc file. Instructions on how to create these are in the github tutorial. Nevertheless, additional resources are listed here.

    -pip install netCDF4 
    
    -pip install xarray
    
    -Create Earth Data login: Information on how this is done can be found at https://www.earthdata.nasa.gov/eosdis/science-system-description/eosdis-components/earthdata-login
    
    -The code will ask to create a netrc file. This file is supposed to have the Earth Data credentials and automate the login into the system. The instructions to create theis file are in the tutorial, but if still confused, more information can be found here: https://nasa-openscapes.github.io/2021-Cloud-Hackathon/tutorials/04_NASA_Earthdata_Authentication.html#authentication-via-netrc-file
    
    -The netrc file should be in the home directory in the computer formatted as such:
    
    machine urs.earthdata.nasa.gov
    login <USERNAME>
    password <PASSWORD>
